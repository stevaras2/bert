```
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```
from keras.models import Model
from keras.layers import Input, Dense
from keras.utils import plot_model

# Multi-layer neural networks
inputs = Input(shape=(3073,))
hidden1 = Dense(64, activation='relu')(inputs)
hidden2 = Dense(128, activation='relu')(hidden1)
hidden3 = Dropout(0.30)(hidden2)
hidden4 = Dense(64, activation='relu')(hidden3)
hidden5 = Dropout(0.25)(hidden4)
hidden6 = Dense(64, activation='relu')(hidden5)
hidden7 = Dropout(0.35)(hidden6)
output = Dense(1, activation='sigmoid')(hidden6)
model = Model(inputs=inputs, outputs=output)


# Model summary
print(model.summary())

# Plot model graph
plot_model(model, show_shapes=True, show_layer_names=True, to_file='model.png')
from IPython.display import Image
Image(retina=True, filename='model.png')
```
